```python
# Step 1: Import
from src.db.maria_client import fetch_orders
from src.model.distance import distance
import pandas as pd

# Step 2: Load data
df = fetch_orders()

# display(df)
# Step 3: Build matrix
ids    = df['id'].tolist()
coords = df[['latitude','longitude']].values
n      = len(ids)

matrix = [
    [distance(coords[i][0], coords[i][1], coords[j][0], coords[j][1]) for j in range(n)]
    for i in range(n)
]

dist_matrix = pd.DataFrame(matrix, index=ids, columns=ids)

# Step 4: Save
dist_matrix.to_csv('../data/dist_matrix.csv', index=True, header=True)

# Step 5: Inspect
dist_matrix
```
